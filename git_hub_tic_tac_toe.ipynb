```python
def checking_bot_win(board):
    flag = 1
    if board[1]+board[2]+board[3]==15 or board[4]+board[5]+board[6]== 15 or board[7]+board[8]+board[9]==15:
        print("bot win")
    elif board[1]+board[4]+board[7]==15 or board[2]+board[5]+board[8]==15 or board[3]+board[6]+board[9]==15:
        print("bot win")
    elif board[1]+board[5]+board[9]==15 or board[3]+board[5]+board[7]==15:
        print("bot win")
    else:
        flag = 0

    if flag==1:
        return True
    return False
```
